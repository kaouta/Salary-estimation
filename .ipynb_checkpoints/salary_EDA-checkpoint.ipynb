```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('salary_data_cleaned.csv')
pd.set_option("display.max_rows", 27)
df
```

```python
def title_simplifier(title):
    if 'data analyst' in title.lower():
        return 'data analyst'
    elif 'business' in title.lower() or 'intelligence' in title.lower():
        return 'business intelligence'
    elif 'data scientist' in title.lower():
        return 'data scientist'
    elif 'Lead'in title.lower():
        return 'Lead data analyst'
    else:
        return 'na'
    
def job_level(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['job_simp']=df['Job Title'].apply(title_simplifier)
df['job_simp'].value_counts()
```

```python
df['experience_level'] = df['Job Title'].apply(job_level)
df.experience_level.value_counts()
```

```python
df['requirements']=df['Job Description'].apply(lambda x:x.replace('\n','').replace('/',''))
df['requirements']
```
