```python
df['salary']=df['Salary Estimate'].apply(lambda x:x.split('(')[0].replace('K',''))
df['min_salary']=df['salary'].apply(lambda x:int(x.split('-')[0].replace('$','')))
df['max_salary']=df['salary'].apply(lambda x:int(x.split('-')[1].replace('$','')))
df['avg_salary']=(df['min_salary']+df['max_salary'])/2
df
```

```python
df['Rating']=df['Rating'].replace(-1,df['Rating'].mean())
df
```

```python
df['State']=df['Location'].str.strip().str[-2:]
df
```

```python
df.State.value_counts()#most offer are in california state
```

```python
df['Company Name']=df['Company Name'].apply(lambda x:x.split('\n')[0])
df
```
